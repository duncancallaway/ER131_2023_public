```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
j_dat = pd.read_csv('JbailyData/data_part2.csv')
j_dat= j_dat.dropna(axis= 0, how='any')
j_dat.describe()
```

```python
plt.scatter(j_dat.poverty, j_dat.pm25);
plt.title('PM2.5 annual concentration versus poverty rate in mid-continent zip codes');
plt.xlabel('Poverty rate')
plt.ylabel('Annual PM2.5, ug/m3')
```

```python
print(j_dat.poverty.shape)
```

```python
x = j_dat.poverty.values.reshape(-1,1)
print(x.shape)
```

```python
j_lm = linear_model.LinearRegression()
j_lm_fit = j_lm.fit(x, j_dat.pm25)
j_dat['y_hat'] = j_lm_fit.predict(x)
```

```python
plt.scatter(x,j_dat.pm25)
plt.plot(x,j_dat.y_hat,'r')
```

```python
e = j_dat.pm25 - j_dat.y_hat
mse = np.sum(e**2)/e.shape[0]
mse
```
